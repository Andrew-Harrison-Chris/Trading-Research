```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
# Creating train and test set
from sklearn.model_selection import train_test_split
X,y = data_all.drop(['Close5','Direction5','Date'],axis=1),data_all['Direction5']

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_inputs = scaler.fit_transform(X)
scaled_inputs = pd.DataFrame(scaled_inputs,columns=X.columns)

X_train,X_test,y_train,y_test = train_test_split(scaled_inputs,y,test_size=.01,random_state=101)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
X_train.head()
```

```python
# Network time
model = Sequential()
model.add(Dense(100, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train,validation_data=(X_test,y_test), epochs=50, batch_size=100, verbose=2)
```
